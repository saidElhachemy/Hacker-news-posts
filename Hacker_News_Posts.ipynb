### data exploring

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
posts = pd.read_csv('HN_posts_year_to_Sep_26_2016.csv')
posts.head()
```

```python
posts.info()
```

```python
posts.describe(include='all')
```

```python
# here we extract the organization/ site from the url
posts.url = posts.url.str.split('/',expand=True)[2]
```

```python
posts.head()
```

```python
# the 50 most frequent sites
posts.url.value_counts().head(50)
```

```python
# the site with the highest sum points is github
posts.groupby('url')['num_points'].sum().reset_index()[posts.groupby('url')['num_points'].sum().reset_index().num_points== posts.groupby('url')['num_points'].sum().reset_index().num_points.max()]
```

```python
# the site with the highest sum comments is nytimes
posts.groupby('url')['num_comments'].sum().reset_index()[posts.groupby('url')['num_comments'].sum().reset_index().num_comments== posts.groupby('url')['num_comments'].sum().reset_index().num_comments.max()]
```

```python
# the user/author with the highest number of posts is jonbaer
posts.groupby('author').count().reset_index()[posts.groupby('author').count().reset_index().id== posts.groupby('author').count().reset_index().id.max()]
```

```python
# the author with the highest number of points is ingve
posts.groupby('author').num_points.sum().reset_index()[posts.groupby('author').num_points.sum().reset_index().num_points== posts.groupby('author').num_points.sum().reset_index().num_points.max()]
```

```python
# the author with the highest number of points is ingve
posts.groupby('author').num_comments.sum().reset_index()[posts.groupby('author').num_comments.sum().reset_index().num_comments== posts.groupby('author').num_comments.sum().reset_index().num_comments.max()]
```

```python
# find the sum of number of points per year
posts.groupby(pd.PeriodIndex(posts['created_at'], freq="Y"))['num_points'].sum().reset_index()
```

```python
# find the sum of number of points per month
posts.groupby(pd.PeriodIndex(posts['created_at'], freq="M"))['num_points'].sum().reset_index()
```

```python
# find the sum of number of points per day
posts_days = posts.groupby(pd.PeriodIndex(posts['created_at'], freq="D"))['num_points'].sum().reset_index()
```

```python
posts_days.num_points.value_counts().sort_index(ascending=False).head(10)
```

```python
# find the day with the highest number of points
posts_days[posts_days.num_points == posts_days.num_points.max()]
```

```python
posts_hightest_day = posts[posts.created_at.str.split(' ',expand=True)[0] == '2/17/2016']
```

```python
posts_hightest_day.num_points.value_counts().sort_index(ascending=False).head(10)
```

```python
# find the post with the highest points
posts_hightest_day[posts_hightest_day.num_points == posts_hightest_day.num_points.max()]
```
